```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']  = (10, 6)
plt.rcParams['axes.labelsize']  = 10
plt.rcParams['axes.titlesize']  = 10
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['lines.linewidth'] = 2

plt.ion()
```

```python
path = "../../results/output.csv"
df = pd.read_csv(path)

df
```

```python
df.groupby("algorithm").describe()
```

```python
df.groupby(["algorithm", "difficulty"])[["expansions"]].mean()
```

```python
df.groupby(["algorithm", "difficulty"])[["expansions", "time"]].describe()
```

```python
x = [*range(1, 9 + 10 + 13 + 1)]
a_expansions = df[df.algorithm == "A"]["expansions"].tolist()
g_expansions = df[df.algorithm == "G"]["expansions"].tolist()
  
plt.plot(x, a_expansions, label = "A* Search") 
plt.plot(x, g_expansions, label = "Greedy Best-first Search")
plt.xlabel("InstÃ¢ncias")
plt.ylabel("# de estados expandidos")
plt.legend()
plt.show()
```

```python
x = [*range(1, 9 + 10 + 13 + 1)]
a_time = df[df.algorithm == "A"]["time"].tolist()
g_time = df[df.algorithm == "G"]["time"].tolist()
  
plt.plot(x, a_time, label = "A* Search") 
plt.plot(x, g_time, label = "Greedy Best-first Search") 
plt.xlabel("InstÃ¢ncias")
plt.ylabel("Tempo (ms)")
plt.legend()
plt.show()
```

```python
b_x = df[df.algorithm == "B"][["clues", "expansions"]].sort_values(by="clues")["clues"].tolist()
b_expansions = df[df.algorithm == "B"][["clues", "expansions"]].sort_values(by="clues")["expansions"].tolist()
u_x = df[df.algorithm == "U"][["clues", "expansions"]].sort_values(by="clues")["clues"].tolist()
u_expansions = df[df.algorithm == "U"][["clues", "expansions"]].sort_values(by="clues")["expansions"].tolist()
i_x = df[df.algorithm == "I"][["clues", "expansions"]].sort_values(by="clues")["clues"].tolist()
i_expansions = df[df.algorithm == "I"][["clues", "expansions"]].sort_values(by="clues")["expansions"].tolist()
g_x = df[df.algorithm == "G"][["clues", "expansions"]].sort_values(by="clues")["clues"].tolist()
g_expansions = df[df.algorithm == "G"][["clues", "expansions"]].sort_values(by="clues")["expansions"].tolist()
a_x = df[df.algorithm == "A"][["clues", "expansions"]].sort_values(by="clues")["clues"].tolist()
a_expansions = df[df.algorithm == "A"][["clues", "expansions"]].sort_values(by="clues")["expansions"].tolist()

plt.plot(b_x, b_expansions, label = "Breadth-first Search") 
plt.plot(u_x, u_expansions, label = "Uniform-cost Search") 
# plt.plot(i_x, i_expansions, label = "Iterative Deepening Search") 
plt.plot(g_x, g_expansions, label = "Greedy Best-first Search") 
plt.plot(a_x, a_expansions, label = "A*") 
plt.xlabel("# de dicas")
plt.ylabel("# de estados expandidos")
plt.legend()
plt.show()
```
